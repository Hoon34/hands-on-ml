> # Decision Tree(Classification) #

#### Like SVM, decision tree is an algorithm which can perform classification, regression and multi-class work. And it is component of random-forest algorithm. ####

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings ; warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
x = iris.data[:,2:]        ## petal length, width
y = iris.target

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=1)
iris_tree.fit(x,y)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize= (8,5))
plot_tree(iris_tree, feature_names= iris.feature_names[2:], class_names= iris.target_names, rounded= True, filled= True)
```

#### We say "root node" which is on top in the image(depth is 0). And a node which doesn't have child node is said leaf node. ####
#### Each node is seperated to minimize impurity calculated with Gini. ####
## $ G_i = 1-\sum_{k=1}^np_{i,k}^2 $ ##

```python
x_seto = x[y==0]
x_versi = x[y==1]
x_virgi = x[y==2]
```

```python
plt.figure(figsize=(10,5)); plt.axis([0,7.3,0,3])
plt.xlabel("Petal length", fontsize=15) ; plt.ylabel("Petal width", fontsize=15)


x0s = np.linspace(0, 7.3, 100)
x1s = np.linspace(0, 3, 100)
x0, x1 = np.meshgrid(x0s, x1s)
X_im = np.c_[x0.flatten(),x1.flatten()]
y_pred = iris_tree.predict(X_im)
plt.contourf(x0, x1, y_pred.reshape(x0.shape), cmap = plt.cm.Pastel2 )


plt.scatter(x_seto[:,0], x_seto[:,1], c='b')
plt.scatter(x_versi[:,0], x_versi[:,1], c='r', marker='s')
plt.scatter(x_virgi[:,0], x_virgi[:,1], c='g', marker='^')
plt.plot([0,8],[0.8,0.8], c='black')
plt.plot([0,8],[1.75,1.75], c='black', linestyle='--')
```

#### Probability that the sample belongs to a class is the ratio of the class in leaf node the sample belonged. ####

```python
print(iris_tree.predict_proba([[3, 1.5]]))
print(iris_tree.predict([[3, 1.5]]))
```

#### CART algorithm is a way of training decision tree. At first, it devide train dataset into two subset using a class k and threshold t. The class and threshold are determined to minimize decision fuction. ####
## $ J(k,t_k) = \frac{m_{\mathsf{left}}}{m}G_{\mathsf{left}} + \frac{m_{\mathsf{right}}}{m}G_{\mathsf{right}} $ ##
#### The argorithm repeats this process until impurity becomes 0 or stop condition is satisfied. ####

#### There are two way getting impurity. One is Gini fuction which we saw above. The other is entrophy. ####
## $ H_i = -\sum_{k=1}^np_{i,k}log_2(p_{i,k}) $ ##
#### Actually there is no difference between two fuction. However, gini is more faster so it is proper as default value. And entrophy makes more balanced tree. ####
#### Some regularization parameter makes a decision tree not fall in overfitting. ####

```python
from sklearn.datasets import make_moons

np.random.seed(123)
X, Y = make_moons(n_samples = 100, noise = 0.2)
x0s = np.linspace(-1.5, 2.5, 100)
x1s = np.linspace(-1, 1.5, 100)
x0, x1 = np.meshgrid(x0s, x1s)
X_im = np.c_[x0.flatten(),x1.flatten()]

tree_1 = DecisionTreeClassifier()
tree_2 = DecisionTreeClassifier(min_samples_leaf=4)

tree_1.fit(X,Y)
tree_2.fit(X,Y)

y_pred1 = tree_1.predict(X_im)
y_pred2 = tree_2.predict(X_im)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (12,5))

ax1.axis([-1.5,2.5, -1, 1.5]); ax1.set_title("No regularization", fontsize=16)
ax1.scatter(X[Y==1][:,0],X[Y==1][:,1], c='red')
ax1.scatter(X[Y==0][:,0],X[Y==0][:,1], c='g')
ax1.contourf(x0, x1, y_pred1.reshape(x0.shape), cmap=plt.cm.binary, alpha=0.2)


ax2.axis([-1.5,2.5, -1, 1.5]); ax2.set_title("min_smaples_leaf = 4", fontsize=16)
ax2.scatter(X[Y==1][:,0],X[Y==1][:,1], c='red')
ax2.scatter(X[Y==0][:,0],X[Y==0][:,1], c='g')
ax2.contourf(x0, x1, y_pred2.reshape(x0.shape), cmap=plt.cm.binary, alpha=0.2)
```
