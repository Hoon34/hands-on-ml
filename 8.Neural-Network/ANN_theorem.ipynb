> # Altificial Neural Network #

ANN is inspired by biological neural networks in our brain.

### 1. Perceptron ###
Perceptron is one of the simplest ANN structures. It is based on TLU (a some different artificial neuron). Input and output values are a number and each input connection is related to weights. TLU calculates linear-combination of input values with weights. And print outcome using step function.

$ h_w(z) = step(z), \quad z = \mathsf{x}^Tw$



A TLU can be used in simple linear binary classification problem. Perceptron consists of TLU with one layer. Each TLU is connected to all inputs. When all neurons in one layer are connected to all neuron in previous layer, we call it fully connected layer or dense layer.



$ h_{W,b}(X)=\phi(XW+b)$

This perceptron has two input and three output values. We can use it to multilabel classification problem. Perceptron is trained using Hebbian learning. In learning process, it reflects errors derived when the network estimates output value.

$ w_{i,j}^{(next \ step)} = w_{i,j} + \mu(y_i-\hat{y}_j)x_i$

### 2. Multilayer Perceptron ###
Multilayer percpetron has an input layer, output layer and at least one hidden TLU layer. All layers have bias neuron (except output layer) and completely connected to next layer.



ANN which has a number of hidden layers is called DNN. Deep learning researches this model. Many researchers had a hard time in finding a method for training MLP. And backpropagation appeared.

One mini-batch is carried out at a time to process the training set. Repeat of this process is called epoch. Each mini-batch is sent to first hidden layer And calculate outputs of all neurons. And the outputs are sent to next layer. This process continues until output of output layer. This is forward pass. Then algorithm estimates output error of network. And calculate degree of contribution to error for each output connection. This algorithm again estimates how much weight of previous layer contributed to contribution degree of the error using chain rule. It continues to arrive at input layer. This backward step efficiently calculates error gradient about all connection weights in network by spreading error gradeint in reverse. At last, algorithm performs gradient descent and modifies all weights in network. 

To make backpropagation work better, some researchers change step fuction to other function. Other two active fuction generally used are ReLU and tanh

$ \mathsf{ReLU}(z)=max(0,z) $

$ tanh(z)=2\sigma(2z)-1 $

### 3. MLP for Regression and Classification ###

In regression work, number of output neuron depends on each problem's output dimension. Generally we don't use active fuction at output neuron to make any value can be derived. However if we want to get a value in certain range, we can use an active fuction.

In binary classification problem, we need one output neuron which has logistic active function. Output value is a value 0(negative) or 1(positive). It is similar to multilabel binary classification. For example, there is a system that classificate an e-mail is spam or not and emergency or not. In this case, we need two ouput neurons. And if each sample belongs to one class among more than 3 classes, we need a neuron to each class. Then we have to use softmax active function at output layer.



### 4. Hyperparameter Tuning ###

- Number of hidden layers

In theory, we can modeling a complex function with one hidden layer which has sufficient neurons. However, in complex problem, DNN is better than shallow neural network. It has better parameter efficiency. Since it uses much less neurons in modeling complex function, It has better score in same train data.

Layer structure helps DNN not only to converge to good solution and to upgrade ability to be generalized at new data. At new neural network, instead initialize first some layers' weights and bias, we can use weights and bias in first layer. Then model doesn't need to learn low-level structure. It is called transfer learning.

- Number of neurons in hidden layer

Number of neurons in input and output layer is determined by shape of input and output. For example, MNIST needs 28X28 input neurons and 10 output neurons.

Composition of hidden layers is like funnel. Neurons in each layer are getting smaller. Because many features in low level can be combined to small features in high level. However in most cases, using same size hidden layers makes same or better performance. Also hyperparameter to tune becomes one.
We can increase number of neurons until overfitting is derived. However in real, choosing a model which has much more layers and neurons and using early stopping and reguralization are simple and efficient.

- Learning rate

Generally best learning rate is half of max learning rate (that train algorithm diverges). One way to search is to start from very low learning rate.
