> # Object Detection

Classificating and estimating location of object is object detection. One method is to classificate an object and find location of object. Then scan the image. Sliding window scans an image and find object. The size of window can be changed bigger or smaller.



This method is easy but it can detect same object several times. So we should remove some unnecessary bounding box. NMS method is widely used.

1. Add objectness class to estimate probability of object's presence. We can train by using sigmoid activation function and binary cross entropy loss. Then delete all bounding box whose score is lower than a threshold.

2. Find the highest score bounding box and delete other bounding box that overlap the bonding box. Repeat this step until there is no bounding box to delete.

 - #### mAP
 
Mean avergae precision is metrics of object detection. Let's figure out recall and precision. These two indicators are trade-off. Then we can draw precision/recall curve. And we can calculate AUC. But area where precision increase with recall increasing is included, especially when recall is little. 

mAP calculates maximum precision from a minimum recall rate of n%. Let's suppose we get maximum precision at minimum recall value 0%, 10%, 20% and 100%. And we get average of the maximum precision values. This is AP. When there are multiple classes, calculate AP about each class and get average of APs. This is mAP.

In object detection, what if system correctly detects right class but location is wrong? We should not include that. One method is to set IOU threshold. For example if IOU is more than 0.5 and class is right, we consider it correct predict. This mAP is written as mAP@0.5 or mAP@50% or mAP@50. Sometimes we should calculate mAP at several IOU threshold. This is average mAP.
