> # Sementic Segmentation

In sementic segmentation, each pixel is classificated to a class that the pixel is involved in like under image. 



Objects that are classificate as same class are not distinguished. The most difficult thing is that image losses location information through CNN. There is a method to solve this problem. Transfer pre-trained CNN into FCN. And this CNN's stride is 32.This means that the last layer prints feature map 32 times smaller than input image. And we adds upsampling layer that increases resolution 32 times.

There are some upsampling methods. One is transpsed convultional layer. This layer injects empty low and column at image and process general convultional step. It can be initailized to work linear interpolation. And since this layer can train, it can be trained well in train set. We can use **Conv2DTranspose**



This still has low accuracy. To overcoem it, adds skip connection. For example, we upsampling ouput image 2 times and adds lower layer's output. Then resolution is two times bigger. Then expand it 16 times. Finally, we achieve 32 times upsampling. This restores some lost location information. We can upsampling bigger than original image.This can be used for increasing image's resolution and called super-resolution.
