> # **Linear Regression** #

- ### Predict of linear Regression ###
#### $$ \hat{y} = w_0x_0 + w_1x_1+\dots+ w_nx_n = W \cdot X \quad (x_0 = 1) $$ ####

- ### Cost function of linear Regression line ###
#### $$ MSE(X,W) = \frac{1}{m} \sum_{i=1}^m (W^TX^i-y^i)^2 $$ ####

#### We want to find a weight vector that has the smallest cost function. ####
#### And there's two methods to find it. ####

### **1. Normal Equation** ###

### $$ \hat{W} = (X^TX)^{-1}X^TY $$ ###
#### **Testing normal equation** ####

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
X = np.random.rand(100,1)
Y = 2 + X * 4 + np.random.randn(100,1)

plt.figure(figsize=(8,5));plt.ylim(-3,10) ; plt.xlabel('X') ; plt.ylabel('Y')
plt.scatter(X,Y)
```

```python
X_cal = np.c_[np.ones((100,1)), X]
W_hat = np.linalg.inv(X_cal.T.dot(X_cal)).dot(X_cal.T).dot(Y)
W_hat
```

#### **We expected 2 and 4. We can not get w hat exactly because of noises, but this is satisfying.** ####

```python
plt.figure(figsize = (8,5));plt.ylim(-3,10) ; plt.xlabel('X') ; plt.ylabel('Y')
plt.scatter(X,Y)
plt.plot( [0, 1], [1.91, 5.95], c='r', linewidth = 2)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, Y)
print(lin_reg.coef_, lin_reg.intercept_)
```

```python
lin_reg.predict([[0],[1]])
```

### **2. Gradient descent** ###

#### Start at w hat which has random value. And that point gradually converges to the minimum point of cost function. ####

### **2-1. Batch gradient descent** ###

#### This method uses all training data for each learing step. In this method, learning rate is important. ####

#### If it's too low, learning progression will takes long time. And if it's too high, it diverges and doesn't converges to best point. ####

```python
def batch_grediant_descent( iteration = 1000, learning_rate = 0.1):
    theta = np.random.randn(2,1)
    for i in range(iteration):
        grad = (2/100) * (X_cal.T.dot(X_cal.dot(theta) - Y))
        theta = theta - learning_rate*grad
    return theta
```

```python
batch_grediant_descent()
```

#### Because this uses all training data for each learning step, It takes long time. ####

### **2-2. Stochastic gradient descent** ###

#### This method uses random sample for each learing step. Therefore, this algorithm is more faster than batch descent and can escape from local minimum more easier. But it's unstable and can't reach at local minimum because it's stochastic. ####

#### To solve this problem, we gradually reduce learning rate. ####

```python
def stochastic_gradient_descent( iteration = 100 ):
    theta = np.random.randn(2,1)
    for i in range(iteration):
        for a in range(100):
            index = np.random.randint(100)
            x1 = X_cal[index:index+1]
            y1 = Y[index:index+1]
            grad = 2 * x1.T.dot(x1.dot(theta) - y1)
            learning = 5 / (i*100 + a + 50)
            theta = theta - learning * grad
    return theta
```

```python
stochastic_gradient_descent()
```

```python
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor(max_iter = 1000, tol = 1e-3, penalty = None, eta0 = 0.1)
sgd.fit(X, Y)
```

```python
print(sgd.intercept_, sgd.coef_)
```
