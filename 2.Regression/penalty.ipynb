> # **Regression with penalty** #

#### To solve overfitting problem, regulating model can be effective solution. Generally, regulation is added to cost function. ####


## **1.Ridge Regression** ##

### $$ J(\theta) = MSE(\theta) + \alpha \frac{1}{2} \sum_{i=1}(\theta_i)^2  $$ ###

### $$ \hat{\theta} = (X^TX + \alpha A)^{-1}X^T y $$ ###

#### Alpha is hyperparameter. User can control how much regulate the model. ####

#### If alpha is 0, ridge regression will be same with linear regression. And if alpha is 1, all weights become to be close at 0. And the model will be mean line. ####

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
X = np.random.randn(30,1)
Y = 0.5 * X + np.random.randn(30,1)
plt.scatter(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

scale = Pipeline([
    ('poly_features', PolynomialFeatures(degree=10, include_bias = False)),
    ('standard', StandardScaler())
])
X_new = scale.fit_transform(X)
```

```python
from sklearn.linear_model import Ridge

ridge_0 = Ridge(alpha = 0, solver = 'cholesky')
ridge_e3 = Ridge(alpha = 1e-03, solver = 'cholesky')
ridge_1 = Ridge(alpha = 1, solver = 'cholesky')
```

```python
x_image = scale.fit_transform(np.arange(-2.5, 1.5, 0.1).reshape(-1,1))
plt.scatter(X,Y, c='b')
for model, color in zip([ridge_0, ridge_e3, ridge_1], ['red', 'lime', 'darkslategrey']):
    model.fit(X_new, Y)
    y_pred = model.predict(x_image)
    plt.plot(np.arange(-2.5, 1.5, 0.1), y_pred, c = color, label = model)
plt.legend()
```

## **2.Lasso Regression** ##

#### This model uses l1 norm of weight vector as penalty. ####
### $$ J(\theta) = MSE(\theta) + \alpha \sum_{i=1}|\theta_i| $$ ###
#### Lasso Regression eliminates some weights which are less important than other weights. So, It automatically selects features and makes sparse model. ####

```python
from sklearn.linear_model import Lasso

lasso_0 = Lasso(alpha = 0)
lasso_e2 = Lasso(alpha = 1e-2)
lasso_1 = Lasso(alpha = 1)
```

```python
x_image = scale.fit_transform(np.arange(-2.5, 1.5, 0.1).reshape(-1,1))
plt.scatter(X,Y, c='b')
for model, color in zip([lasso_0, lasso_e2, lasso_1], ['red', 'lime', 'darkslategrey']):
    model.fit(X_new, Y)
    y_pred = model.predict(x_image)
    plt.plot(np.arange(-2.5, 1.5, 0.1), y_pred, c = color, label = model)
plt.legend()
```

## **3.Elastic net Regression** ##

#### This model's shape is that ridge and lasso are mixed. The penalty is just adding l1 and l2 penalty. And the ratio is determined by r. ####
### $$ J(\theta) = MSE(\theta) + r\alpha \sum_{i=1}|\theta_i|+ \frac{1-r}{2}\alpha \sum_{i=1}(\theta_i)^2 $$ ###

```python
from sklearn.linear_model import ElasticNet
elastic = ElasticNet(alpha = 0.1, l1_ratio = 0.5)
elastic.fit(X_new,Y)
```

```python
y_pred = elastic.predict(x_image)
plt.scatter(X,Y, c='b')
plt.plot(np.arange(-2.5, 1.5, 0.1), y_pred, c = 'r', label = 'Elastic_net')
plt.legend(loc = 'upper left')
```

#### Then what algorithm should we choose among simple linear, ridge, lasso and elastic net? ####

#### In most cases, some penalty is needed to make effective model. So ridge can be basic model. And if it seems that only few features are used, lasso and elastic are useful. But when the number of features are bigger than the number of train samples, elastic is prefered because of lasso's problem. ####
