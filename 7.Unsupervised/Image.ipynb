> # Imange Segmentation Using Cluster
#### In sementic segmentation, all pixels that belongs to same object are assigned to same segment. We will do color segmentation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings("ignore")
```

```python
from matplotlib.image import imread
image = imread("bori.jpg")
image.shape
```

```python
plt.figure(figsize=(6,6))
plt.imshow(image)
```

```python
from sklearn.cluster import KMeans

x = image.reshape(-1,3)
kmeans = KMeans(n_clusters=8).fit(x)
segment_bori = kmeans.cluster_centers_[kmeans.labels_]
segment_bori = segment_bori.reshape(image.shape)
```

```python
plt.figure(figsize=(6,6))
plt.title("8 colors", fontsize=15)
plt.imshow(segment_bori.astype(np.int))
```

```python
plt.figure(figsize=(8,8))
plt.axis(False)
for i, n in zip([10, 8, 6, 4], [221,222,223,224]):
    kmeans = KMeans(n_clusters=i).fit(x)
    km_image = kmeans.cluster_centers_[kmeans.labels_]
    km_image = km_image.reshape(image.shape).astype(np.int)
    plt.subplot(n)
    plt.imshow(km_image)
    plt.title("{} Color".format(i))
    plt.axis("off")
```
