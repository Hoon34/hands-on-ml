## 1. K-means ##

```python
from sklearn.datasets import make_blobs
center = [[0.5, 2], [-0.5,2], [-1,1], [0,1], [1,1]]
noise = [0.2,0.1, 0.4, 0.3, 0.3]
```

```python
x,y = make_blobs(n_samples=1000, centers= center, cluster_std= noise, random_state=4)
plt.scatter(x[:,0], x[:,1] , s=5)
```

```python
from sklearn.cluster import KMeans
k = 5
kmeans = KMeans(n_clusters=k)
kmeans.fit(x)
```

#### Hyperparameter **k** is number of clusters. In this example, we can set **k** after watching data image. But choosing k is not easy. ####

```python
y_pred = kmeans.predict(x)
y_pred[:10]
```

#### Each sample is assinged to one of five clusters.(0 to 5) And we can also check centroids of clusters. ####

```python
kmeans.cluster_centers_
```

```python
x0s = np.linspace(-2, 2, 500)
x1s = np.linspace(0, 2.7, 270)
x0s, x1s = np.meshgrid(x0s, x1s)
x_im = np.c_[x0s.flatten(), x1s.flatten()]
y_pred = kmeans.predict(x_im)

plt.axis([-2, 2, 0, 2.7])
plt.contourf(x0s, x1s, y_pred.reshape(x0s.shape), cmap = plt.cm.Pastel2, alpha=0.7)
plt.scatter(x[:,0], x[:,1], s=5, c="black")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c="r")
```

#### Some samples are assigned to wrong label. K-means algorithmn doesn't work well when size clusters are different. This method also provides distance from each label.(Soft clustering) ####

```python
kmeans.transform([[1.5, 1]])
```

#### At first, centriods are setted randomnly. Next, samples are assinged to labels and centriods are updated. This progress repeats until centroids are no more changed. ####

```python
kmeans_iter1 = KMeans(n_clusters=5, init="random", n_init=1, max_iter=1, algorithm="full", random_state=5)
kmeans_iter2 = KMeans(n_clusters=5, init="random", n_init=1, max_iter=2, algorithm="full", random_state=5)
kmeans_iter3 = KMeans(n_clusters=5, init="random", n_init=1, max_iter=3, algorithm="full", random_state=5)

md1 = kmeans_iter1.fit(x)
md2 = kmeans_iter2.fit(x)
md3 = kmeans_iter3.fit(x)

ypred1 = md1.predict(x_im)
ypred2 = md2.predict(x_im)
ypred3 = md3.predict(x_im)
```

```python
fig, axes = plt.subplots(3,2, figsize=(12,10))

for pred , i in zip((ypred1, ypred2), (0,1)):
    axes[i][1].contourf(x0s, x1s, pred.reshape(x0s.shape), cmap = plt.cm.Pastel2, alpha=0.9)
    axes[i+1][0].contourf(x0s, x1s, pred.reshape(x0s.shape), cmap = plt.cm.Pastel2, alpha=0.9)
axes[2][1].contourf(x0s, x1s, ypred3.reshape(x0s.shape), cmap = plt.cm.Pastel2, alpha=0.9)

for md, i in zip((md1, md2, md3), (0,1,2)):
    axes[i][0].scatter(x[:,0], x[:,1], s=2, c="black")
    axes[i][0].scatter(md.cluster_centers_[:,0], md1.cluster_centers_[:,1], c="r")
    axes[i][1].scatter(x[:,0], x[:,1], s=2, c="black")
    axes[i][1].scatter(md.cluster_centers_[:,0], md1.cluster_centers_[:,1], c="r")
    
axes[0][0].set_title("Centeriod Updating", fontsize=15)
axes[0][1].set_title("Label Assinging", fontsize=15)
```

#### Centroid setting is ramdom so different result can be derived. ####

```python
sol1 = KMeans(n_clusters=5, init="random", n_init=1, algorithm="full", random_state=10)
sol2 = KMeans(n_clusters=5, init="random", n_init=1, algorithm="full", random_state=13)

sol1.fit(x)
sol2.fit(x)

sol1_y = sol1.predict(x_im)
sol2_y = sol2.predict(x_im)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))

for ax, sol in zip((ax1, ax2), (sol1_y, sol2_y)):
    ax.contourf(x0s, x1s, sol.reshape(x0s.shape), cmap = plt.cm.Pastel2, alpha=0.9)
    ax.scatter(x[:,0], x[:,1], c="black", s=2)

ax1.scatter(sol1.cluster_centers_[:,0], sol1.cluster_centers_[:,1], c="r")
ax2.scatter(sol2.cluster_centers_[:,0], sol2.cluster_centers_[:,1], c="r")
```

#### Another method is to implement algorithmn multiply with different random initailizaion and select the best solution. **n_init** controls number of random initialization.(default is 10.) Scikit-learn returns the best solution among them. ####
#### The performance score of k-means algorithm is mean-squared-distance from the closest centriod. It is called inerteria of model. And score is negative value of inerteria.(Big value is good. - rule of scikit-learn) ####

```python
kmeans.inertia_
```

```python
kmeans.score(x)
```

#### K-means ++ algorithmn reduces the probability of converging to none-best solution. ####
#### 1.Choose one centroid $c^{(i)}$ from dataset randomnly and uniformly. ####
#### 2.Selected sample $\mathrm{x}^{(i)}$ to new centriod $c^{(i)}$ with probability of $D(\mathrm{x}^{(i)})^2/\sum^m_{j=1}D(\mathrm{x}^{(j)})^2. \ D(\mathrm{x}^{(j)})$ is distance between sample $\mathrm{x}^{(i)}$ and selected the closested centriod. ####
#### 3.Repeat previous step until $k$ centriods are selected. ####

### -Minibatch K-means ###
#### This algorithmn uses mini batch for each repeat. It improves speed of algorithm 3-4 times higher. ####

```python
from sklearn.cluster import MiniBatchKMeans
mini_kmeans = MiniBatchKMeans(n_clusters=5)
mini_kmeans.fit(x)
```

```python
from timeit import timeit
```

```python
k_inertia = []
mini_inertia =[]
k_time = []
mini_time = []

for i in range(50):
    km = KMeans(n_clusters=i+1)
    minikm = MiniBatchKMeans(n_clusters=i+1)
    a = timeit("km.fit(x)", number=10, globals=globals())
    b = timeit("minikm.fit(x)", number=10, globals=globals())
    
    k_inertia.append(km.inertia_)
    k_time.append(a)
    mini_inertia.append(minikm.inertia_)
    mini_time.append(b)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.title("inertia", fontsize=15)
plt.axis([0,50,0,100])
plt.plot(range(1,51), k_inertia, c='r', linestyle="--", label="k-means")
plt.plot(range(1,51), mini_inertia, c='blue', label="minibatch k-means")
plt.legend(fontsize=12)

plt.subplot(122)
plt.title("train time(s)", fontsize=15)
plt.axis([0,50,0,4])
plt.plot(range(1,51), k_time, c='r', linestyle="--", label="k-means")
plt.plot(range(1,51), mini_time, c='blue', label="minibatch k-means")
```

#### Generaly, We doesn't know how many clusters we should set. Inertia cannot be an indicator for setting the number. In above image, inertia gets smaller as k gets bigger. ####

```python
plt.axis([1,8.5,0,600])
plt.plot(range(1,10), k_inertia[:9], marker='o')
```

#### More accurate method is silhouette score. It is mean of silhouette coefficient about all samples. Silhouette coefficient is calculated $(b-a)/max(a,b)$. $a$ is mean distance from other samples in same clusters. $b$ is mean distance from the nearest cluster. Siluhouette coefficient can be changed between -1 to 1. As this indicator is bigger, the sample is assigned to right cluster more accurately. ####

```python
from sklearn.metrics import silhouette_score

silou = []
for i in range(1,10):
    md = MiniBatchKMeans(n_clusters=i+1)
    md.fit(x)
    silou.append(silhouette_score(x, md.labels_))

plt.figure(figsize=(8,4))
plt.axis([1.9, 8.5, 0.4, 0.6])
plt.plot(range(2,11), silou, marker='o')
```

#### This graph shows 5 is good choice. ####

#### K-means has some limits. We should run the algorithmn to avoid not optimal solution. And we should choose number of clusters. Also, it doesn't work well in different size of clusters and not spherical clusters. ####

```python
X1, y1 = make_blobs(n_samples=1000, centers=((4, -4), (0, 0)), random_state=42)
X1 = X1.dot(np.array([[0.374, 0.95], [0.732, 0.598]]))
X2, y2 = make_blobs(n_samples=250, centers=1, random_state=42)
X2 = X2 + [6, -8]
X = np.r_[X1, X2]
y = np.r_[y1, y2]

plt.scatter(X[:,0], X[:,1], s=4)
```

```python
good = KMeans(n_clusters=3, init=np.array([[0,0], [-1.5,2], [3.5,1]]), n_init=1)
bad = KMeans(n_clusters=3)
good.fit(X)
bad.fit(X)
```

```python
x0s = np.linspace(-5, 7, 500)
x1s = np.linspace(-4, 7, 500)
x0s, x1s = np.meshgrid(x0s, x1s)
x_im = np.c_[x0s.flatten(), x1s.flatten()]

sol1_y = good.predict(x_im)
sol2_y = bad.predict(x_im)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))

for ax, sol in zip((ax1, ax2), (sol1_y, sol2_y)):
    ax.contourf(x0s, x1s, sol.reshape(x0s.shape), cmap = plt.cm.Pastel2, alpha=0.9)
    ax.scatter(X[:,0], X[:,1], c="black", s=2)

ax1.scatter(good.cluster_centers_[:,0], good.cluster_centers_[:,1], c="r")
ax2.scatter(bad.cluster_centers_[:,0], bad.cluster_centers_[:,1], c="r")
```

## 2.DBSCAN ##
#### This algorithmn defines a dense area as a cluster. ####
#### 1. Algorithmn counts how many sample is within $\epsilon$. This area is called $\epsilon$-neighborhood.
#### 2. If there are at least min_samples number samples in $\epsilon$-neighborhood, it is regarded as core instance.
#### 3. All samplese around core instance belongs to same cluster. Other core instances can belong to the same cluster. And they continuously form a cluster.
#### 4. Neither core instance nor neighbor sample is regarded as outlier.

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

x,y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_[:5]
```

```python
plt.scatter(x[:,0], x[:,1], c=dbscan.labels_, cmap=plt.cm.rainbow)
plt.title("eps=0.05", fontsize=15)
```

#### DBSCAN has only 2 hyperparameter epsilon and min_sample. Through controling them properly, we can get good clusters. ####

```python
dbscan2 = DBSCAN(eps=0.2, min_samples=5)
dbscan2.fit(x)
plt.scatter(x[:,0], x[:,1], c=dbscan2.labels_, cmap=plt.cm.Dark2)
plt.title("eps=0.2", fontsize=15)
```

#### But this method doesn't provide "predict()" method, only "fit_predict()" method. It means this algorithmn can't predict cluster about new sample. ####
