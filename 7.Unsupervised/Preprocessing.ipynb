> # Preprocessing using clustering #

We can use cluster for process preprocessing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_digits
X, y = load_digits(return_X_y= True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
```

```python
log.score(X_test, y_test)
```

We get about 96% accuracy from general logistic regression.

```python
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=55)),
    ("logis", LogisticRegression())
])
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

The accuracy has imporved a little. It means even though clusters reduced dimension of dataset to 50, reduced data can be classificated linearly better. We randomly set cluster number $k$. let's search best $k$ using gridsearch.

```python
from sklearn.model_selection import GridSearchCV

param = dict(kmeans__n_clusters = range(2,100))
grid = GridSearchCV(pipeline, param, cv=3, verbose=2)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.score(X_test, y_test)
```
