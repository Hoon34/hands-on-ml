# **Classification(MNIST Dataset)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
import warnings

warnings.filterwarnings('ignore')
jtplot.style('monokai', grid = False)
```

```python
train_df = pd.read_csv('train.csv', header = 0)
train_df.shape
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(train_df, test_size = 0.2)
```

```python
train.head()
```

```python
x_train = train.drop(['label'], axis = 1 )
y_train = train['label']
x_test = test.drop(['label'], axis = 1 )
y_test = train['label']
```

```python
an_image = x_train.iloc[3].values.reshape(28,28)
plt.imshow(an_image, cmap = 'binary')
print(y_train.iloc[3])
```

**Binary Classification( 5 or not )**

```python
y_train_5 = (y_train == 5 )
y_test_5 = (y_test == 5)
```

```python
from sklearn.linear_model import SGDClassifier
clf_5 = SGDClassifier()
clf_5.fit(x_train, y_test_5)
```

```python
print(clf_5.predict( x_test.iloc[3].values.reshape(1,-1)), y_test.iloc[3])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf_5, x_train, y_train_5, scoring = 'accuracy')
```

 A model that predicts all images are 'not 5' has at least 90% accuracy score.
 (because only about 10% images are 5.)

-> Use **confusion matrix** 



**Precision = TP / (TP + FP)**

**Recall = TP/ (TP + FN)**

**F1 score = 2/ (1/Precision + 1/Recall)**

```python
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(clf_5, x_train, y_train_5)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_train_5, y_train_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
print('Precision:', precision_score(y_train_5, y_train_pred))
print('Recall:', recall_score(y_train_5, y_train_pred))
print('F1 Score:', f1_score(y_train_5, y_train_pred))
```

**Trade-off between precision and recall**

```python
#If decision function is larger than threshold, output is True.
#And the threshold 
score = clf_5.decision_function(x_train.iloc[3].values.reshape(1,-1))
score
```

```python
y_scores = cross_val_predict(clf_5, x_train, y_train_5, method = 'decision_function')
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, threshold = precision_recall_curve(y_train_5, y_scores)
```

```python
plt.xlim(-50000,50000); plt.ylim(0,1.02)
plt.xlabel('Threshold')
plt.grid(True)

plt.plot(threshold, precision[:-1], linestyle = '--', c='red', label = 'Precision')
plt.plot(threshold, recall[:-1], linestyle = '-', c='lime', label = 'Recall')
plt.plot([threshold[32000],threshold[32000]], [recall[32000],precision[32000]],
         c='white', marker='o', linestyle = '--')
plt.legend(fancybox = True, shadow = True)
```

**As threshhold in model is changed, Trade-off between precicion and recall is watched.**

**In low recall score, high precision score is useless**

```python
plt.xlim(0,1);plt.ylim(0,1); plt.grid(True); plt.xlabel('Recall'); plt.ylabel('Precision')

plt.plot(recall[:-1], precision[:-1], c='lime')
plt.plot([-1,recall[32000],recall[32000]], [precision[32000], precision[32000], -1], 
        c= 'red', linestyle= '--', marker = 'o')
```

```python
## Find threshold that makes precision at least 0.9

threshold_90_precision = threshold[np.argmax(precision >= 0.9)]
y_train_pred = (y_scores>= threshold_90_precision)
print( 'Precision:', precision_score(y_train_5, y_train_pred) )
print( 'Recall:', recall_score(y_train_5, y_train_pred) )
```

**ROC Curve is also useful tool in binary classification**

**It uses FPR(at x) and TPR(at y, recall)**

**A model which has the largest area under the curve(AUC) is the best model**

```python
from sklearn.metrics import roc_curve

fpr, tpr, threshold = roc_curve(y_train_5, y_scores)
```

```python
plt.figure(figsize=(10,5)); plt.title('ROC Curve'); plt.grid(True)
plt.xlim(0,1);plt.ylim(0,1); plt.xlabel('FPR'); plt.ylabel('TPR')

plt.plot(fpr,tpr, c='red', linewidth = 2)
plt.plot([0,1], [0,1], linestyle = '--', c = 'white')
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5, y_scores)
```

**Find another model which has better score.(randomforest)**

```python
from sklearn.ensemble import RandomForestClassifier

rforest_clf = RandomForestClassifier()
y_prob_forst = cross_val_predict(rforest_clf, x_train, y_train_5, 
                                 method = 'predict_proba')
```

```python
y_prob_forst[:3]
```

```python
y_prob_true = y_prob_forst[:,1]
fpr_forest, tpr_forest, threshold_forest = roc_curve(y_train_5, y_prob_true)
```

```python
plt.figure(figsize = (10,5)); plt.grid(True)
plt.xlim(0,1); plt.ylim(0,1.01); plt.xlabel('FPR'); plt.ylabel('TPR'); plt.title('ROC')

plt.plot(fpr,tpr, c ='red', linewidth = 2, linestyle ='--', label = 'SGD')
plt.plot(fpr_forest,tpr_forest, c = 'lime', linewidth = 2, label = 'RandomForest')
plt.plot([0,1],[0,1],c = 'white', linestyle = '--')
plt.legend(loc = 'lower right')
```

```python
print('SGD:', roc_auc_score(y_train_5, y_scores))
print('RandomForest: ', roc_auc_score(y_train_5, y_prob_true))
```

**RandomForest Classifier model has better score than SGD model.**
