> # RNN

Recurrent neural netwrok can handle sequence data. This network can predict what will happen in the future. It can be stock price, autonomous-driving of a vehicle. It can get arbitrary length of sequence not fixed length.

- #### Recurrent neuron and layer

Main different point is RNN has circulating connection at back. This recurrent neuron has $\mathbf{x}_{(t)}$ and $y_{(t-1)}$ at each time step $t$. At first time step, $y_{(t-1)}$ is set 0. All input and output become vectors. We can also unrolling the network through time.



Eacn reccurent neuron has 2 weights. One is for input $\mathbf{x}_{(t)}$ and the other is for previous output $y_{(t-1)}$. Suppose these weight vectors as $\mathbf{w}_{(x)}$ and $\mathbf{w}_{(y)}$. If we think all neural network, we can change vectors into weight metrices $\mathbf{W}_{(x)}$ and $\mathbf{W}_{(y)}$. Then output vector of all reccurent layer is calculated as under fomular.

$y_{(t)} = \phi (\mathbf{W}_{x}^T\mathbf{x}_{(t)} \ + \ \mathbf{W}_{y}^T \mathbf{y}_{(t-1)} \ + \mathbf{b})$

And we can calcualte all output of reccurent layer by making all inputs metrix $\mathbf{X}_{(t)}$

$\mathbf{Y}_{(t)} = \phi (\mathbf{X}_{(t)}\mathbf{W}_{x} \ +\mathbf{Y}_{(t-1)}\mathbf{W}_{y} \ + \mathbf{b})$

$\quad \ \  = \phi([\mathbf{X}_{(t)}\mathbf{Y}_{(t-1)}]\mathbf{W}\ + \mathbf{b}) \quad \quad \quad $   in $\mathbf{W} = $ $\begin{bmatrix}\mathbf{W}_x\\ \mathbf{W}_y\\ \end{bmatrix}$

$\mathbf{Y}_{(t)}$ is function of $\mathbf{X}_{(t)}$ and $\mathbf{Y}_{(t-1)}$ and $\mathbf{Y}_{(t-1)}$ is function of $\mathbf{X}_{(t-1)}$ and $\mathbf{Y}_{(t-2)}$. So $\mathbf{Y}_{(t)}$ becomes function of all inputs from t=0.

Output of reccurent neuron is a function of all inputs of previous time step so it can be considered memory shape. Memory cell is element of neural network that preserves a state through time step. One reccurent neuron or layer is a basic cell that can train short pattern.

State of cell $\mathbf{h}_{(t)}$ is a function of input of that time step and state of previous time step. So, $\mathbf{h}_{(t)} = f(\mathbf{h}_{(t-1)}, \mathbf{x}_{(t)})$. 

RNN get input sequence and can make output sequence. This is sequence-to-sequence network. It is useful at prediction of stock price. Or we can ignore all ouput except last value. This is sequence-to-vector network. It is useful at sentiment analysis. In reverse, we can repeatly inject one input vector at each time step and print a sequence. This is vector-to-sequence network. At alst, we can connect sequence-to-vector encoder network and vector-to-sequence decoder network. This works well in translation.
