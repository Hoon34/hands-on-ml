```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings ; warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('iris.csv', header= 0)
data['Species_num'] = data['Species'].replace(['Iris-setosa', 'Iris-versicolor','Iris-virginica'], [1,2,3])
data
```

```python
X = data[['PetalLengthCm', 'PetalWidthCm']]
Y = data['Species_num']
```

```python
from sklearn.svm import LinearSVC

x_im = X.loc[(Y == 1) + (Y == 2)]
y_im = Y.loc[(Y == 1) + (Y == 2)]

im_model = LinearSVC()
im_model.fit(x_im,y_im)

print(im_model.coef_, im_model.intercept_)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,4))

ax1.axis((0, 5.5, 0, 2)); ax1.set_xlabel('PetalLength') ; ax1.set_ylabel('PetalWidth')
ax1.scatter(data[Y==1]['PetalLengthCm'], data[Y==1]['PetalWidthCm'], marker = 'x', c = 'g', label = 'Iris setosa')
ax1.scatter(data[Y==2]['PetalLengthCm'], data[Y==2]['PetalWidthCm'], marker = 's', c='r', label = 'Iris versicolor')
ax1.plot([2,4],[0,4]) ; ax1.plot([0,6],[1,0.5])
ax1.legend()

ax2.axis((0, 5.5, 0, 2)) ; ax2.set_xlabel('PetalLength')
ax2.scatter(data[Y==1]['PetalLengthCm'], data[Y==1]['PetalWidthCm'], marker = 'x', c = 'g', label = 'Iris setosa')
ax2.scatter(data[Y==2]['PetalLengthCm'], data[Y==2]['PetalWidthCm'], marker = 's', c='r', label = 'Iris versicolor')

x1 = np.arange(0,6,0.5) ; y1 = (-0.61*x1 + 2.13)/0.81 ; y2 = (-0.61*x1 + 2.13 + 0.6)/0.81 ; y3 = (-0.61*x1 + 2.13 - 0.6)/0.81
ax2.plot(x1,y1,c='black') ; ax2.plot(x1,y2,c='black',linestyle = '--') ; ax2.plot(x1,y3,c='black',linestyle = '--')
```

#### There are 2 linear classification model in left image. It seems that they classificate the iris data well. However, decision function of two model is too close to the sample. So maybe they cannot estimate well about samples close to decision function. ####

#### And there is SVM linear classification model in right image. It perfectly classificate the iris data and the decision function is located as far as possible from the samples. We can think that SVM model is finding a road whose width is largest between the classes. The decision boundary is not affected by samples outside the boundary. Adding a new sample outside the boundary does not affect the decision boundary. ####

```python
from sklearn.preprocessing import StandardScaler
im_2 = pd.DataFrame([[1, 50, 1],
                    [5, 23, 1],
                    [3, 79, 2],
                    [5, 55, 2]], columns=['x0', 'x1', 'class'])
x_im_2 = im_2[['x0', 'x1']]
y_im_2 = im_2['class']
scaler = StandardScaler()
x_im_2_scaled = scaler.fit_transform(x_im_2)
```

```python
y_im_2
```

```python
im_2_model = LinearSVC(C=1000)
im_2_model_scaled = LinearSVC()
im_2_model.fit(x_im_2, y_im_2)
im_2_model_scaled.fit(x_im_2_scaled, y_im_2)
```

```python
print(im_2_model.coef_, im_2_model.intercept_)
print(im_2_model_scaled.coef_, im_2_model_scaled.intercept_)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,3))

ax1.axis((0, 6, 0, 100))
ax1.scatter(x_im_2.iloc[:2,0], x_im_2.iloc[:2,1], marker = 's', c = 'g')
ax1.scatter(x_im_2.iloc[2:,0], x_im_2.iloc[2:,1], c='r')
x1 = np.arange(0,10,3) ; y1 = (0.11*x1 + 0.44) / 0.03
ax1.plot(x1, y1, c='black')

ax2.axis((-2, 2, -2, 2)) ; ax1.set_xlabel('x0') ; ax1.set_ylabel('x1')
ax2.scatter(x_im_2_scaled[:2,0], x_im_2_scaled[:2,1], marker = 's', c = 'g')
ax2.scatter(x_im_2_scaled[2:,0], x_im_2_scaled[2:,1], c='r')
x1 = np.arange(-3,3,1) ; y1 = (-0.54*x1 ) / 0.86
ax2.plot(x1, y1, c='black')
```

#### SVM model is sensitive to scale of features. Right image shows difference scale of features can derive error of model. So proper scaler is needed to increase score of model. ####

## - Soft mragin classification ##

#### If all samples are classificated outside road, we call it hard margin classification. This algorithm has 2 problems. Data should be classified linearly and model is sensitive to outliers. To avoid this problems, We need more flexible model. model should set balance between width of road and margin violation. And this model is soft margin classification. ####

#### There are many several hyperparameters in the model. **C** is one of them. When C is low, we get a model like left image.(allow several margin violation) And when C is high, we get model like right image.(allow a few margin violation) ####
