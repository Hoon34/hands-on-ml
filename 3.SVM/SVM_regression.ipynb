> # SVM Regression #

#### We can use SVM algorithm in linear, non-linear regression problem. The method to solve regression problem is thinking the other way around. SVM regression let many samples located in margin error as many as possible. ####

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings ; warnings.filterwarnings('ignore')
```

```python
x = np.random.randn(70)
y = 1.5*x + 2 + 4 * np.random.rand(70)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.svm import LinearSVR
reg1 = LinearSVR(epsilon=1)
reg2 = LinearSVR(epsilon=0.5)
```

```python
x_reg = x.reshape(-1,1)
reg1.fit(x_reg,y)
reg2.fit(x_reg,y)
```

```python
print(reg1.coef_, reg1.intercept_)
print(reg2.coef_, reg2.intercept_)
```

```python
fig , (ax1, ax2) = plt.subplots(1,2, figsize=(12,5))

a = np.array([-2,2])
a1 = a*reg1.coef_ + reg1.intercept_
b0 = a1 + 1
b1 = a1 - 1
ax1.set_title("epsilon = 1", fontsize = 15)
ax1.scatter(x,y)
ax1.plot(a, a1 ,c='black')
ax1.plot(a, b0 ,c='black', linestyle = '--')
ax1.plot(a, b1 ,c='black', linestyle = '--')


a2 = a*reg2.coef_ + reg2.intercept_
c0 = a2 + 0.5
c1 = a2 - 0.5
ax2.set_title("epsilon = 0.5", fontsize = 15)
ax2.scatter(x,y)
ax2.plot(a, a2 ,c='black')
ax2.plot(a, c0 ,c='black', linestyle = '--')
ax2.plot(a, c1 ,c='black', linestyle = '--')
```
